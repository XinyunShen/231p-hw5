```python
import matplotlib.pyplot as plt 
file_name = "output.txt"
f = open(file_name, 'r')
single = []
coarse_grain = []
for line in f:
    line = line.split('\n')[0]
    line = line.split(',')
    if (len(line) != 3):
        continue
    single.append(float(line[0]))
    coarse_grain.append(float(line[2]))

x = range(10,10001,100)
plt.plot(x, single,'r-',label='single thread')
plt.plot(x, coarse_grain, 'b-', label='coarse-grain multi-thread')
plt.xlabel('matrix size')
plt.ylabel('runtime (seconds)')
plt.legend()
plt.size(80,100)
plt.show()

    
```
